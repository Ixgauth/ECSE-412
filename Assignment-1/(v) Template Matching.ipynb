```python
import cv2
import numpy as np
import skimage.measure
import matplotlib.pyplot as plt

#import image and template then flip rgb values
image_path = "where_is_waldo.jpg"

template_path = "template1.png"

image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

template = cv2.imread(template_path)
template = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)

#duplicate the image so it can be displayed later
img_dup = image.copy()

#get the height and width of the template so we can draw rectangle later
ht, wt = template.shape[0:2]

#run the sum of squared difference on the image and template
res = cv2.matchTemplate(image, template, cv2.TM_SQDIFF)

#print the values of the sum of squared difference as asked for in project description
print(res);

#find the location and value of both the closest point to the template and the farthest from the template
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res);

#print the location and value of the point where we found Waldo
print(min_loc, " ", min_val);

#draw the rectangle around waldo
cv2.rectangle(img_dup, min_loc, (min_loc[0]+wt, min_loc[1]+ht), (255,255,0), 5)

#print the original image and the image with Waldo boxed in
plt.figure(figsize=(20,20))
plt.subplot(121), plt.imshow(image)
plt.title("Template Image"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_dup)
plt.title("Matched Image"), plt.xticks([]), plt.yticks([])
plt.show()
```

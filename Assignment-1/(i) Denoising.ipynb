```python
import cv2
import numpy as np
import skimage.measure
import matplotlib.pyplot as plt

#intake both of the needed images
image_path = "peppers.png"
image_path_2 = "gaussian_noise_peppers.png"

image1 = cv2.imread(image_path)
image2 = cv2.imread(image_path_2)

#flip the images rgb values
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

#perform gaussian filtering
res1 = cv2.GaussianBlur(image2, (7,7), 2)

#perform box filtering
kernel = np.ones((7,7), np.float32)/(7*7)
res2 = cv2.filter2D(image2, -1, kernel)

#calculate the PSNR of both images with the image not containing noise
psnr1 = skimage.measure.compare_psnr(image1, res1)
psnr2 = skimage.measure.compare_psnr(image1, res2)

#plot the images created by both types of of filtering
plt.figure(figsize=(15, 15))
plt.subplot(121), plt.imshow(res1)
plt.title("Gaussian Filtered"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(res2)
plt.title("Box Filtered"), plt.xticks([]), plt.yticks([])

plt.show()

print ('The difference in PSNR between the clean image and the gaussian filter is: ', psnr1, '\n')
print('The difference in PSNR between the clean image and the box filter is: ', psnr2, '\n')
```

Thus the gaussian filtered image has a higher PSNR than the box filtered one. Thus it has given a superior result

```python
#import the two neeeded images
image_path = "peppers.png"
image_path_2 = "saltandpepper_noise_peppers.png"

image1 = cv2.imread(image_path)
image2 = cv2.imread(image_path_2)

#convert both images to correct RGB
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

#perform gaussian filter
res1 = cv2.GaussianBlur(image2, (7,7), 2)

#perform median filter
res2 = cv2.medianBlur(image2, 7)

#check PSNR of both images with the image not containing noise
psnr1 = skimage.measure.compare_psnr(image1, res1)
psnr2 = skimage.measure.compare_psnr(image1, res2)

#plot the images created by both types of filtering
plt.figure(figsize=(15, 15))
plt.subplot(121), plt.imshow(res1)
plt.title("Gaussian Filtered"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(res2)
plt.title("Median Filtered"), plt.xticks([]), plt.yticks([])

plt.show()

print ('The difference in PSNR between the clean image and the gaussian filter is: ', psnr1, '\n')
print('The difference in PSNR between the clean image and the median filter is: ', psnr2, '\n')
```

Thus the median filtered image has a higher PSNR than the gaussian filtered one. Therefore it is has done a better job of denoising the image
